## **stack, unstack**
 - stack : 컬럼 레벨에서 인덱스 레벨로 데이터프레임을 변경합니다.
 - unstack : 인덱스 레벨에서 컬럼 레벨로 데이터프레임을 변경합니다.
  - stack의 반대입니다.

```python
import pandas as pd

file_path = '/content/drive/MyDrive/data/titanic_train.csv'
df = pd.read_csv(file_path)
```

```python
df.head(5)
```

```python
pivot = pd.pivot_table(df, index=['Sex','Pclass'], values=['Survived','Fare'],
                       aggfunc=['mean','median','sum'])
pivot

#피벗테이블로 다중인덱스, 다중컬럼을 만들어본거임
#인덱스,컬럼 모두 level이 2임
```

```python
pivot.stack(0) #컬럼의 첫번째 레벨을 인덱스로 내립니다.
# 첫번째 컬럼을 인덱스로 내리는 이것을 stack이라고함
#stack(0) stack함수 안에 0은 몇번째 level의 컬럼을 내릴건지에 대한 숫자임
#0을 넣어서 첫번째 레벨을 인덱스로 낼니거
```

```python
pivot.stack(1) #컬럼의 두번째 레벨을 인덱스로 내립니다.
```

```python
pivot.unstack(0) #인덱스의 첫번째 레벨을 컬럼으로 쌓아 올립니다

#첫번째 인덱스가 성별이였는데 성별이 컬럼으로 바뀐것을 볼 수 있음
```

```python
pivot.unstack(1) #인덱스의 두번째 레벨을 컬럼으로 쌓아 올립니다
#두번째 인덱스였던 pclass가 컬럼으로 올라간것을 볼 수 있지
```

## **melt**
- `pd.melt(데이터명, id_vars=기준 컬럼)`
![이미지](https://i0.wp.com/cmdlinetips.com/wp-content/uploads/2019/06/Pandas_melt_reshape.png?w=421&ssl=1)

```python
data = pd.DataFrame({'name':['a','b','c']
                    , 'order_count':[3,4,10]
                    , 'amount':[10000,25000,300000]})
```

```python
data
```

```python
pd.melt(data, id_vars=['name'])

#Wide 데이터를 long데이터로 바뀜
# 기준되는 name 컬럼으로
```

```python
pd.melt(data, id_vars=['name'], var_name='type', value_name='val')
#컬럼의 이름을 바로 바꿀 수도 있음
```
